### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 10, 100)
y = np.sin(x)
axes[0, 0].plot(x, y, color='red', linewidth=2)
axes[0, 0].set_title("Line Plot")
axes[0, 0].set_xlabel("X-axis")
axes[0, 0].set_ylabel("Y-axis")
axes[0, 0].grid(True)


categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['red', 'black', 'red', 'black'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# 3rd Subplot: Scatter Plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='black', marker='o')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X-axis")
axes[1, 0].set_ylabel("Y-axis")
axes[1, 0].grid(True)

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['orange', 'white', 'green', 'blue']
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
axes[1, 1].set_title("Pie Chart")

plt.tight_layout()


plt.show()
```
